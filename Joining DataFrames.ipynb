```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({"key":["a","b","c","c","d","e"],
                      "num1":range(6)})
df2=pd.DataFrame({"key":["b","c","e","f"],
                  "num2":range(4)})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2) #merges the common datas
```

```python
pd.merge(df1,df2, on="key")
```

```python
df3=pd.DataFrame({"key1":["a","b","c","c","d","e"],
                      "num1":range(6)})
df4=pd.DataFrame({"key2":["b","c","e","f"],
                  "num2":range(4)})
print(df3)
print(df4)
```

```python
pd.merge(df3,df4,left_on="key1", right_on="key2") #underdtand this as union
```

```python
pd.merge(df1, df2, how="outer") #all the value are shown
```

```python
pd.merge(df1, df2, how="left")  #left ko table intact
```

```python
pd.merge(df1, df2, how="right")  #merge from right garda right ko table intact
```

```python
pd.merge(df1, df2, how="inner")   #intersection
```

```python
right=pd.DataFrame([[1,2],[3,4],[5,6]],
                  index=["a","c","d"],
                  columns=["Tom","Tim"])
left=pd.DataFrame([[7,8],[9,10],[11,12],[13,14]],
                  index=["a","b","e","f"],
                  columns=["Sam","Kim"])

print(right)
print(left)
```

```python
pd.merge(right, left,
        right_index=True,
        left_index=True,
        how="outer")
```

```python
left.join(right)
```

```python
data=pd.DataFrame([[1,2],[5,7],[9,11]],index=["a","b","f"],columns=["Alex","Keta"])
data
```

```python
left.join([right, data]) #inner join by default
```

```python
#airline with high delay deaprture
delayed_flights=flights[flights['dep_delay']>0].groupby('carrier').size().reset_index(name="num_delay")
delayed_flights
```

```python
delayed_flights.merge(airlines,on="carrier", how="left")
```

```python
flights = flights[flights['dep_delay'] > 0] # filter for delays
carrier_delay_count = flights.groupby('carrier').size().reset_index(name='num_delay') # count the delays by carrier
max_delay_carrier = carrier_delay_count[carrier_delay_count['num_delay'] == carrier_delay_count['num_delay'].max()] # find the carrier with the most delays
most_delayed_airline = pd.merge(max_delay_carrier, airlines, on='carrier', how='left')['name'] # join with airlines table and select the name of the most delayed airline
```

```python
most_delayed_airline
```

```python
flights.groupby('dest').agg(mean_delay=('arr_delay','mean'))
```

```python
flights.groupby('dest').agg(mean_delay=('arr_delay'
                                        , 'mean')).reset_index().query('mean_delay == mean_delay.min()')
```

```python
flights.groupby('dest').agg(mean_delay=('arr_delay', 'mean')).reset_index()\
    .query('mean_delay == mean_delay.min()').merge(airports, left_on='dest', right_on='faa', how='left')
```

```python
flights.groupby('dest').agg(mean_delay=('arr_delay', 'mean')).reset_index()\
    .query('mean_delay == mean_delay.min()')\
    .merge(airports, left_on='dest', right_on='faa')\
    .loc[:, ['name']]
```
