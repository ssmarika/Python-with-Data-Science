```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= anscombe.iloc[:,1]
y= anscombe.iloc[:,5]
print(x)
print(y)
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
a=2.0
b=0.7
#calculate residuals
residuals=y-(a+b*x)
print(residuals)

#craete scatter plot with regression line
plt.scatter(x,y)
plt.plot(x, a+b *x, color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from matplotlib.patches import Rectangle
```

```python
b= 0.7

#assume nagari, actual value nikali garna lai:
a= np.mean(y)-b*np.mean(x)
residuals= (a+b*x)-y

fig, ax= plt.subplots()
ax.scatter(x,y,marker='s',s=100)
ax.plot(x, a+b*x, color='red')
for i in range(len(x)):
    rect= Rectangle((x[i],y[i]),
                     residuals[i],
                     residuals[i],
                     fill= True,
                     color='green',
                     alpha=0.5)
    ax.add_patch(rect)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(3,17)
ax.set_ylim(3,12)
plt.show()
```
