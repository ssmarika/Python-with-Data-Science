```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
df=pd.DataFrame(np.random.rand(12).reshape(6,2), columns=["x", "y"]
               )
df
```

```python
x= df.iloc[:,0]
y= df.iloc[:,1]
print(x)
print(y)
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
a=0.02
b=0.7
#calculate residuals
residuals=y-(a+b*x)
print(residuals)

#craete scatter plot with regression line
plt.scatter(x,y)
plt.plot(x, a+b *x, color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

b=np.sum((x-np.mean(x))*(y-np.mean(y)))/np.sum((x-np.mean(x))**2)
a= np.mean(y)-b*np.mean(x)
residuals= (a+b*x)-y

fig, ax= plt.subplots()
ax.scatter(x,y,marker='s',s=100)
ax.plot(x, a+b*x, color='red')
for i in range(len(x)):
    rect= Rectangle((x[i],y[i]),
                     residuals[i],
                     residuals[i],
                     fill= True,
                     color='green',
                     alpha=0.5)
    ax.add_patch(rect)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(0.0,1.0)
ax.set_ylim(0.0,0.7)
plt.show()
```
